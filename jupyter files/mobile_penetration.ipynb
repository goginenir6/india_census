```python
# How does the mobile penetration vary in regions (districts or states) with high or low
# agricultural workers?

import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt

#read data from csv file
df=pd.read_csv('india-districts-census-2011.csv')
filt = df.groupby(['State name'], as_index=False)[['Agricultural_Workers', 'Households_with_Telephone_Mobile_Phone']].sum()
# filt

# filt['State name']
n_groups = len(filt['State name'])
Agricultural_Workers = filt['Agricultural_Workers']
Households_with_Telephone = filt['Households_with_Telephone_Mobile_Phone']
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, Agricultural_Workers, bar_width, 
                 alpha=opacity,
                 color='b',
                 label='Agricultural Workers')

 
rects2 = plt.bar(index + bar_width, Households_with_Telephone, bar_width, 
                 alpha=opacity,
                 color='g',
                 label='Households with Telephone Mobile Phone')
 

plt.xlabel('States')
plt.ylabel('Agriculture vs Mobile Usage In 10 millions')
plt.title('Mobile enetration vary in states with Agricultural Workers')
plt.xticks(index + bar_width, filt['State name'], rotation='vertical', fontsize='7')
# plt.yticks(index + bar_width, Agricultural_Workers)
plt.legend()

#set y axis values manually
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,0,30000000))



#shows y values on top of bars. But its not looks good.
# def autolabel(rects):
#     """
#     Attach a text label above each bar displaying its height
#     """
#     for rect in rects:
#         height = rect.get_height()
#         ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
#                 '%d' % int(height),
#                 ha='center', va='bottom', rotation='vertical')

# autolabel(rects1)
# autolabel(rects2

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
 
plt.tight_layout()
plt.show()


```
