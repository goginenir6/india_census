```python
import pandas as pd
import numpy as np
import itertools


df=pd.read_csv('india-districts-census-2011.csv')
first_st = df[df['State name'] =='BIHAR' ]
second_st = df[df['State name'] =='TAMIL NADU']

i =0
matched_cols = []
for val in first_st.values:
    dist_code = val[0]
    for col, val in itertools.izip(first_st.columns,val):
        for sub in second_st.values:
            dist_code1 = sub[0]
            for col1, val1 in itertools.izip(second_st.columns,sub):
                if (col== col1) and (val == val1):
                    print str(i)+'   '+str(dist_code)+', '+str(dist_code1)+', '+str(col)+', '+str(val)+', '+str(val1)
#                     matched_cols.append({'BH_dist_code': dist_code, 'TM_dist_cde':dist_code1, 'Matched_coloumn':col, 'BH_val':val,'TM_val': val1})
                    matched_cols.append((dist_code,dist_code1, col,val,val1))
                    i = i+1
lables = ['BH_dist_code', 'TM_dist_cde', 'Matched_col_name', 'BH_val','TM_val']
result = pd.DataFrame(matched_cols, columns=lables)
result
        
```
